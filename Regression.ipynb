   3    One Hot encoding

```python
new_df = df.copy()
new_df
```

```python
colums_names = list(new_df.select_dtypes(include=['object']))
```

```python
df_fin = pd.get_dummies(new_df, columns=colums_names, drop_first=True).copy()
df_fin
```

    df_fin

```python
X = df_fin[[name for name in df_fin.columns if name != 'price']]
y = df_fin['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
y_test
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(data=reg.coef_, index=X_train.columns, columns=['Coefficient'])
coeff_df
```

```python
y_pred = reg.predict(X_test)
```

```python
print(f'training r-squared {reg.score(X_train, y_train)}')
print(f'testing r-squared {reg.score(X_test, y_test)}')
```

```python
pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
